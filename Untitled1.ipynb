```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score # Keep accuracy metric
import bloqade
from bloqade.ir.location import Chain, start
from time import time # Keep time for potential timing if needed




# Set random seed for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# Load CIFAR-10 dataset with custom preprocessing
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# Load datasets
train_dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
test_dataset = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform)

# Manual grayscale conversion using NumPy
train_X = np.mean(train_dataset.data, axis=3)  # Average across color channels
test_X = np.mean(test_dataset.data, axis=3)

# Normalize to 0-1 range
train_X = train_X.astype(np.float32) / 255.0
test_X = test_X.astype(np.float32) / 255.0
train_y = np.array(train_dataset.targets)
test_y = np.array(test_dataset.targets)

# Hyperparameters
dim_ae = 8  # Latent dimension for Autoencoder
num_train_examples = 1000  # Limit the training set
num_test_examples = 200  # Limit the test set
num_epochs_ae = 500
batch_size_ae = 64
input_dim = train_X.shape[1] * train_X.shape[2]  # Flattened image dimension

# Reshape images for Autoencoder
train_X_reshaped = train_X.reshape(len(train_X), -1)
test_X_reshaped = test_X.reshape(len(test_X), -1)

# Prepare data for Autoencoder
train_X_ae = torch.tensor(train_X_reshaped[:num_train_examples], dtype=torch.float32)
test_X_ae = torch.tensor(train_X_reshaped[num_train_examples:num_train_examples+num_test_examples], dtype=torch.float32)

# Autoencoder Model
class Autoencoder(nn.Module):
    def __init__(self, input_dim, latent_dim):
        super(Autoencoder, self).__init__()
        self.encoder = nn.Sequential(
            nn.Linear(input_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, latent_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, input_dim),
            nn.Sigmoid()  # Output in 0-1 range
        )

    def forward(self, x):
        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return encoded, decoded

# Initialize Autoencoder
input_dim = train_X_reshaped.shape[1]
latent_dim = dim_ae
autoencoder = Autoencoder(input_dim, latent_dim)

# Loss and Optimizer
criterion_ae = nn.MSELoss()
optimizer_ae = optim.Adam(autoencoder.parameters(), lr=0.001, weight_decay=1e-5)

# Create datasets
train_dataset_ae = TensorDataset(train_X_ae, train_X_ae)
val_dataset_ae = TensorDataset(test_X_ae, test_X_ae)

train_loader_ae = DataLoader(train_dataset_ae, batch_size=batch_size_ae, shuffle=True)
val_loader_ae = DataLoader(val_dataset_ae, batch_size=batch_size_ae, shuffle=False)

# Training loop with validation
train_losses = []
val_losses = []

for epoch in range(num_epochs_ae):
    # Training phase
    autoencoder.train()
    train_epoch_loss = 0.0
    for batch_data, _ in train_loader_ae:
        optimizer_ae.zero_grad()
        _, decoded = autoencoder(batch_data)
        loss_ae = criterion_ae(decoded, batch_data)
        loss_ae.backward()
        optimizer_ae.step()
        train_epoch_loss += loss_ae.item()
    
    # Validation phase
    autoencoder.eval()
    val_epoch_loss = 0.0
    with torch.no_grad():
        for batch_data, _ in val_loader_ae:
            _, decoded = autoencoder(batch_data)
            val_loss = criterion_ae(decoded, batch_data)
            val_epoch_loss += val_loss.item()
    
    # Average losses
    train_loss = train_epoch_loss / len(train_loader_ae)
    val_loss = val_epoch_loss / len(val_loader_ae)
    
    # Store losses
    train_losses.append(train_loss)
    val_losses.append(val_loss)
    
    # Print progress
    if (epoch+1) % 50 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs_ae}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')

# Visualization of reconstructions
def visualize_reconstructions(autoencoder, test_data, num_images=5):
    autoencoder.eval()
    with torch.no_grad():
        _, reconstructed = autoencoder(test_data[:num_images])
        
    fig, axes = plt.subplots(2, num_images, figsize=(15, 6))
    for i in range(num_images):
        # Original images
        axes[0, i].imshow(test_data[i].numpy().reshape(32, 32), cmap='gray')
        axes[0, i].set_title('Original')
        axes[0, i].axis('off')
        
        # Reconstructed images
        axes[1, i].imshow(reconstructed[i].numpy().reshape(32, 32), cmap='gray')
        axes[1, i].set_title('Reconstructed')
        axes[1, i].axis('off')
    
    plt.tight_layout()
    plt.show()

# Plot loss curves
plt.figure(figsize=(10, 5))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.title('Autoencoder Loss Curves')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Visualize reconstructions
visualize_reconstructions(autoencoder, test_X_ae)

# Get encoded representations
autoencoder.eval()
with torch.no_grad():
    xs_ae_torch, _ = autoencoder(train_X_ae)
    xt_ae_torch, _ = autoencoder(test_X_ae)
    xs_ae = xs_ae_torch.numpy()
    xt_ae = xt_ae_torch.numpy()

# Scaling the AE components
spectral_ae = np.amax(xs_ae) - np.amin(xs_ae)
m1_ae = np.amin(xs_ae)
xs_ae_scaled = (xs_ae - m1_ae) / spectral_ae
xs_test_ae_scaled = (xt_ae - m1_ae) / spectral_ae

print("Min AE = ", np.min(xs_ae_scaled))
print("Max AE = ", np.max(xs_ae_scaled))


QRC_parameters = {
    "atom_number": dim_ae, # Use dim_ae instead of dim_pca
    "geometry_spec": Chain(dim_ae, lattice_spacing=10), # Use dim_ae instead of dim_pca
    "encoding_scale": 15.0,
    "rabi_frequency": 3.0,
    "total_time": 6,
    "time_steps": 8,
    "readouts": "ZZ",
}

# Function to build Bloqade tasks (local detuning) (No changes needed)
def build_task_local(QRC_parameters, xs1):
    natoms=QRC_parameters["atom_number"]
    encoding_scale=QRC_parameters["encoding_scale"]
    dt=QRC_parameters["total_time"]/QRC_parameters["time_steps"]
    rabi_oscillations_program = (QRC_parameters["geometry_spec"]
        .rydberg.rabi.amplitude.uniform.constant(
            duration="run_time", value=QRC_parameters["rabi_frequency"]
        )
        .detuning.uniform.constant(duration="run_time", value=encoding_scale/2)
        .scale(list(xs1)).constant(duration="run_time", value=-encoding_scale)
        )
    rabi_oscillation_job = rabi_oscillations_program.batch_assign(run_time=np.arange(1, QRC_parameters["time_steps"]+1, 1)*dt)
    return rabi_oscillation_job

# Function to build Bloqade tasks (global detuning) (No changes needed)
def build_task_global(QRC_parameters, xs1):
    natoms=QRC_parameters["atom_number"]
    encoding_scale=QRC_parameters["encoding_scale"]
    dt=QRC_parameters["total_time"]/QRC_parameters["time_steps"]
    rabi_oscillations_program = (QRC_parameters["geometry_spec"]
        .rydberg.rabi.amplitude.uniform.constant(
            duration="run_time", value=QRC_parameters["rabi_frequency"]
        )
        .detuning.uniform.constant(duration="run_time", value=encoding_scale/2)
        .constant(duration="run_time", value=-encoding_scale * np.sum(xs1))
        )
    rabi_oscillation_job = rabi_oscillations_program.batch_assign(run_time=np.arange(1, QRC_parameters["time_steps"]+1, 1)*dt)
    return rabi_oscillation_job

# Function to build Bloqade tasks (hybrid detuning) (No changes needed)
def build_task_hybrid(QRC_parameters, xs1):
   natoms=QRC_parameters["atom_number"]
   encoding_scale=QRC_parameters["encoding_scale"]
   dt=QRC_parameters["total_time"]/QRC_parameters["time_steps"]
   rabi_oscillations_program = (QRC_parameters["geometry_spec"]
        .rydberg.rabi.amplitude.uniform.constant(
            duration="run_time", value=QRC_parameters["rabi_frequency"]
        )
        .detuning.uniform.constant(duration="run_time", value=encoding_scale/2)
        .constant(duration="run_time", value=-encoding_scale*np.sum(xs1)/natoms)
        .scale(list(xs1)).constant(duration="run_time", value=-encoding_scale/natoms)

        )
   rabi_oscillation_job = rabi_oscillations_program.batch_assign(run_time=np.arange(1, QRC_parameters["time_steps"]+1, 1)*dt)
   return rabi_oscillation_job


# Function to process simulation results (No changes needed)
def process_results(QRC_parameters, report):
    embedding = []
    natoms = QRC_parameters["atom_number"]
    try:
        for t in range(QRC_parameters["time_steps"]):
            ar1 = -1.0 + 2 * ((report.bitstrings())[t])
            nsh1 = ar1.shape[0]
            for i in range(natoms):
                embedding.append(np.sum(ar1[:, i]) / nsh1)
            if QRC_parameters["readouts"] == "ZZ":
                for i in range(natoms):
                    for j in range(i + 1, natoms):
                        embedding.append(np.sum(ar1[:, i] * ar1[:, j]) / nsh1)
    except:
        print("No results exist.")
        for t in range(QRC_parameters["time_steps"]):
            for i in range(natoms):
                embedding.append(0.0)
            if QRC_parameters["readouts"] == "ZZ":
                for i in range(natoms):
                    for j in range(i + 1, natoms):
                        embedding.append(0.0)
    return embedding

# Function to process sample results (No expectation) (No changes needed)
def process_results_samples(QRC_parameters, report):
  embedding=[]
  try:
      embedding=report.bitstrings()
  except:
      print("No results exist.")
      for t in range(QRC_parameters["time_steps"]):
          for i in range(natoms):
              embedding.append(0.0)
          if QRC_parameters["readouts"]=="ZZ":
              for i in range(natoms):
                  for j in range(i+1,natoms):
                      embedding.append(0.0)
  return embedding

# Function to get embeddings using Bloqade simulation (local) (No changes needed)
def get_embeddings_emulation_local(xs, num_examples, nshots=10000):
    return np.array([
        process_results(
            QRC_parameters,
            build_task_local(QRC_parameters, xs[data, :])
            .bloqade.python()
            .run(shots=nshots, rtol=1e-8, atol=1e-8)
            .report(),
        )
        for data in range(num_examples)
    ])
# Function to get embeddings using Bloqade simulation (global) (No changes needed)
def get_embeddings_emulation_global(xs, num_examples, nshots=10000):
    return np.array([
        process_results(
            QRC_parameters,
            build_task_global(QRC_parameters, xs[data, :])
            .bloqade.python()
            .run(shots=nshots, rtol=1e-8, atol=1e-8)
            .report(),
        )
        for data in range(num_examples)
    ])

# Function to get embeddings using Bloqade simulation (hybrid) (No changes needed)
def get_embeddings_emulation_hybrid(xs, num_examples, nshots=10000):
    return np.array([
        process_results(
            QRC_parameters,
            build_task_hybrid(QRC_parameters, xs[data, :])
            .bloqade.python()
            .run(shots=nshots, rtol=1e-8, atol=1e-8)
            .report(),
        )
        for data in range(num_examples)
    ])



#####################################################################
# 4. Generate QRC Embeddings (Using AE-reduced data for QRC)
#####################################################################
embeddings_local = get_embeddings_emulation_local(xs_ae_scaled, num_train_examples, nshots=10000) # Use xs_ae_scaled
test_embeddings_local = get_embeddings_emulation_local(xs_test_ae_scaled, num_test_examples, nshots=10000) # Use xs_test_ae_scaled
embeddings_global = get_embeddings_emulation_global(xs_ae_scaled, num_train_examples, nshots=10000) # Use xs_ae_scaled
test_embeddings_global = get_embeddings_emulation_global(xs_test_ae_scaled, num_test_examples, nshots=10000) # Use xs_test_ae_scaled
embeddings_hybrid = get_embeddings_emulation_hybrid(xs_ae_scaled, num_train_examples, nshots=10000) # Use xs_ae_scaled
test_embeddings_hybrid = get_embeddings_emulation_hybrid(xs_test_ae_scaled, num_test_examples, nshots=10000) # Use xs_test_ae_scaled


#Plot the QRC Embeddings mean value (No changes needed)
mean_embed = np.mean(embeddings_local, axis=0)
mean_test_embed = np.mean(test_embeddings_local, axis=0)
plt.figure(figsize=(10, 5))
plt.plot(mean_embed, label = "QRC Local Training Embeddings")
plt.plot(mean_test_embed, label="QRC Local Testing Embeddings")
plt.title("Mean QRC Embeddings Local")
plt.xlabel("Entry")
plt.ylabel("Mean")
plt.legend()
plt.show()

mean_embed = np.mean(embeddings_global, axis=0)
mean_test_embed = np.mean(test_embeddings_global, axis=0)
plt.figure(figsize=(10, 5))
plt.plot(mean_embed, label = "QRC Global Training Embeddings")
plt.plot(mean_test_embed, label="QRC Global Testing Embeddings")
plt.title("Mean QRC Embeddings Global")
plt.xlabel("Entry")
plt.ylabel("Mean")
plt.legend()
plt.show()

mean_embed = np.mean(embeddings_hybrid, axis=0)
mean_test_embed = np.mean(test_embeddings_hybrid, axis=0)
plt.figure(figsize=(10, 5))
plt.plot(mean_embed, label = "QRC Hybrid Training Embeddings")
plt.plot(mean_test_embed, label="QRC Hybrid Testing Embeddings")
plt.title("Mean QRC Embeddings Hybrid")
plt.xlabel("Entry")
plt.ylabel("Mean")
plt.legend()
plt.show()

print("Embeddings (local):", embeddings_local.shape)
print("Test Embeddings (local):", test_embeddings_local.shape)
print("Embeddings (global):", embeddings_global.shape)
print("Test Embeddings (global):", test_embeddings_global.shape)
print("Embeddings (hybrid):", embeddings_hybrid.shape)
print("Test Embeddings (hybrid):", test_embeddings_hybrid.shape)


#####################################################################
# 5. PyTorch Models (Simplified CNN)
#####################################################################

# Linear Model in PyTorch (No changes needed)
class LinearModel(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearModel, self).__init__()
        self.linear = nn.Linear(input_size, output_size)

    def forward(self, x):
        return self.linear(x)

# Simple Neural Network in PyTorch (No changes needed)
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size) # Reduced to one hidden layer
        #self.relu2 = nn.ReLU() # Removed second ReLU and FC layer
        #self.fc3 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        #x = self.relu2(self.fc2(x)) # Removed second layer
        x = self.fc2(x) # Changed to fc2 directly
        return x

# Simplified CNN Class
class CNN(nn.Module):
    def __init__(self, input_size, output_size):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv1d(in_channels = 1, out_channels= 8, kernel_size = 3, padding = 1) # Reduced filters, smaller kernel
        self.relu1 = nn.ReLU()
        self.pool1 = nn.MaxPool1d(kernel_size = 2)
        #self.conv2 = nn.Conv1d(in_channels = 16, out_channels = 32, kernel_size = 5, padding = 2) # Removed second Conv layer
        #self.relu2 = nn.ReLU()
        #self.pool2 = nn.MaxPool1d(kernel_size = 2)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(input_size // 2 * 8, output_size) # Adjusted input size for FC layer

    def forward(self, x):
      x = x.unsqueeze(1)
      x = self.relu1(self.conv1(x))
      x = self.pool1(x)
      #x = self.relu2(self.conv2(x)) # Removed second Conv layer
      #x = self.pool2(x)
      x = self.flatten(x)
      x = self.fc1(x)
      return x

#####################################################################
# 6. Training/Testing Utility Functions (No changes needed)
#####################################################################
def train_and_test_pytorch(model, train_data, train_targets, test_data, test_targets, num_epochs, batch_size, learning_rate):
        # Convert data to tensors
        train_data = torch.tensor(train_data, dtype=torch.float32)
        train_targets = torch.tensor(train_targets, dtype=torch.int64)
        test_data = torch.tensor(test_data, dtype=torch.float32)
        test_targets = torch.tensor(test_targets, dtype=torch.int64)

        #Data load for batched operations
        train_dataset = TensorDataset(train_data, train_targets)
        train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
        test_dataset = TensorDataset(test_data, test_targets)
        test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

        # Loss function and optimizers
        criterion = nn.CrossEntropyLoss()
        optimizer = optim.Adam(model.parameters(), lr=learning_rate, eps=0.0002)


        # Training loop
        losses = []
        accuracies = []
        for epoch in range(num_epochs):
            model.train()
            for batch_data, batch_targets in train_loader:
                optimizer.zero_grad()  # Reset gradients
                outputs = model(batch_data)
                loss = criterion(outputs, batch_targets)
                loss.backward() # Calculate gradients
                optimizer.step() # Update weights
                losses.append(loss.item())

            # Evaluation of the accuracy
            model.eval()  # Switch model to evaluation mode
            correct = 0
            total = 0
            with torch.no_grad():  # Do not track gradients during evaluation
                for batch_data, batch_targets in test_loader:
                    outputs = model(batch_data)
                    _, predicted = torch.max(outputs.data, 1)
                    total += batch_targets.size(0)
                    correct += (predicted == batch_targets).sum().item()
            accuracies.append(100 * correct / total)

        test_accuracy = 100 * correct / total
        return test_accuracy, losses, accuracies

#####################################################################
# 7. Training and Testing (Autoencoder Baseline)
#####################################################################

# Linear Classification on AE data using PyTorch
model_ae = LinearModel(input_size=xs_ae_scaled.shape[1], output_size=10) # Use xs_ae_scaled
ae_accuracy, ae_losses, ae_accuracies = train_and_test_pytorch(
        model_ae,
        xs_ae_scaled, # Use xs_ae_scaled
        train_y[:num_examples],
        xs_test_ae_scaled, # Use xs_test_ae_scaled
        test_y[:num_test_examples],
        num_epochs=1000,
        batch_size=100,
        learning_rate=0.1
    )
print('PyTorch AE test accuracy (Linear Model):', round(ae_accuracy, 2), "%")

# Linear Classification on QRC embeddings using PyTorch (Local)
model_qrc_local = LinearModel(input_size=embeddings_local.shape[1], output_size=10)
qrc_accuracy_local, qrc_losses_local, qrc_accuracies_local = train_and_test_pytorch(
        model_qrc_local,
        embeddings_local,
        train_y[:num_examples],
        test_embeddings_local,
        test_y[:num_test_examples],
        num_epochs=2000,
        batch_size=100,
        learning_rate=0.0002
    )
print('PyTorch QRC test accuracy (Local Linear):', round(qrc_accuracy_local, 2), "%")


# Linear Classification on QRC embeddings using PyTorch (Global)
model_qrc_global = LinearModel(input_size=embeddings_global.shape[1], output_size=10)
qrc_accuracy_global, qrc_losses_global, qrc_accuracies_global = train_and_test_pytorch(
    model_qrc_global,
    embeddings_global,
    train_y[:num_examples],
    test_embeddings_global,
    test_y[:num_test_examples],
    num_epochs=2000,
    batch_size=100,
    learning_rate=0.0002
)
print('PyTorch QRC test accuracy (Global Linear):', round(qrc_accuracy_global, 2), "%")

# Linear Classification on QRC embeddings using PyTorch (Hybrid)
model_qrc_hybrid = LinearModel(input_size=embeddings_hybrid.shape[1], output_size=10)
qrc_accuracy_hybrid, qrc_losses_hybrid, qrc_accuracies_hybrid = train_and_test_pytorch(
    model_qrc_hybrid,
    embeddings_hybrid,
    train_y[:num_examples],
    test_embeddings_hybrid,
    test_y[:num_test_examples],
    num_epochs=2000,
    batch_size=100,
    learning_rate=0.0002
)
print('PyTorch QRC test accuracy (Hybrid Linear):', round(qrc_accuracy_hybrid, 2), "%")


# Simple NN on AE Data
model_nn_ae = SimpleNN(input_size=xs_ae_scaled.shape[1], hidden_size=50, output_size=10) # Use xs_ae_scaled
nn_accuracy_ae, nn_losses_ae, nn_accuracies_ae = train_and_test_pytorch(
        model_nn_ae,
        xs_ae_scaled, # Use xs_ae_scaled
        train_y[:num_examples],
        xs_test_ae_scaled, # Use xs_test_ae_scaled
        test_y[:num_test_examples],
        num_epochs=1000,
        batch_size=100,
        learning_rate=0.001
)
print('PyTorch Simple NN test accuracy (AE):', round(nn_accuracy_ae, 2), "%")

# CNN on AE data
model_cnn_ae = CNN(input_size = xs_ae_scaled.shape[1] , output_size=10) # Use xs_ae_scaled
cnn_accuracy_ae, cnn_losses_ae, cnn_accuracies_ae = train_and_test_pytorch(
        model_cnn_ae,
        xs_ae_scaled, # Use xs_ae_scaled
        train_y[:num_examples],
        xs_test_ae_scaled, # Use xs_test_ae_scaled
        test_y[:num_test_examples],
        num_epochs=2000,
        batch_size=100,
        learning_rate=0.001
)
print('PyTorch CNN test accuracy (AE):', round(cnn_accuracy_ae, 2), "%")


# CNN on QRC data (Local)
model_cnn_local = CNN(input_size = embeddings_local.shape[1] , output_size=10)
cnn_accuracy_local, cnn_losses_local, cnn_accuracies_local = train_and_test_pytorch(
        model_cnn_local,
        embeddings_local,
        train_y[:num_examples],
        test_embeddings_local,
        test_y[:num_test_examples],
        num_epochs=2000,
        batch_size=100,
        learning_rate=0.001
)
print('PyTorch CNN test accuracy (Local):', round(cnn_accuracy_local, 2), "%")
# CNN on QRC data (Global)
model_cnn_global = CNN(input_size = embeddings_global.shape[1] , output_size=10)
cnn_accuracy_global, cnn_losses_global, cnn_accuracies_global = train_and_test_pytorch(
        model_cnn_global,
        embeddings_global,
        train_y[:num_examples],
        test_embeddings_global,
        test_y[:num_test_examples],
        num_epochs=2000,
        batch_size=100,
        learning_rate=0.001
)
print('PyTorch CNN test accuracy (Global):', round(cnn_accuracy_global, 2), "%")
# CNN on QRC data (Hybrid)
model_cnn_hybrid = CNN(input_size = embeddings_hybrid.shape[1] , output_size=10)
cnn_accuracy_hybrid, cnn_losses_hybrid, cnn_accuracies_hybrid = train_and_test_pytorch(
        model_cnn_hybrid,
        embeddings_hybrid,
        train_y[:num_examples],
        test_embeddings_hybrid,
        test_y[:num_test_examples],
        num_epochs=2000,
        batch_size=100,
        learning_rate=0.001
)
print('PyTorch CNN test accuracy (Hybrid):', round(cnn_accuracy_hybrid, 2), "%")

#####################################################################
# 10. Plotting (Updated to remove SVM and include AE results)
#####################################################################

# Plot training losses for the neural nets
plt.figure(figsize=(10, 5))
plt.plot(ae_losses, label="AE Linear Model") # AE Linear model
plt.plot(pca_losses, label="PCA Linear Model") # PCA linear model for comparison, if you kept pca_losses
plt.plot(qrc_losses_local, label="QRC Linear Model (local)")
plt.plot(qrc_losses_global, label="QRC Linear Model (Global)")
plt.plot(qrc_losses_hybrid, label="QRC Linear Model (Hybrid)")
plt.plot(nn_losses_ae, label="Simple NN (AE)") # Simple NN on AE
plt.plot(nn_losses, label="Simple NN (PCA)") # Simple NN on PCA, if you kept nn_losses
plt.plot(cnn_losses_ae, label = "CNN (AE)") # CNN on AE
plt.plot(cnn_losses_local, label = "CNN (Local QRC)")
plt.plot(cnn_losses_global, label = "CNN (Global QRC)")
plt.plot(cnn_losses_hybrid, label = "CNN (Hybrid QRC)")
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()

# Plot training accuracies for the neural nets
plt.figure(figsize=(10, 5))
plt.plot(ae_accuracies, label="AE Linear Model") # AE Linear model
plt.plot(pca_accuracies, label="PCA Linear Model") # PCA linear model for comparison, if you kept pca_accuracies
plt.plot(qrc_accuracies_local, label="QRC Linear Model (local)")
plt.plot(qrc_accuracies_global, label="QRC Linear Model (Global)")
plt.plot(qrc_accuracies_hybrid, label="QRC Linear Model (Hybrid)")
plt.plot(nn_accuracies_ae, label="Simple NN (AE)") # Simple NN on AE
plt.plot(nn_accuracies, label="Simple NN (PCA)") # Simple NN on PCA, if you kept nn_accuracies
plt.plot(cnn_accuracies_ae, label = "CNN (AE)") # CNN on AE
plt.plot(cnn_accuracies_local, label = "CNN (Local QRC)")
plt.plot(cnn_accuracies_global, label = "CNN (Global QRC)")
plt.plot(cnn_accuracies_hybrid, label = "CNN (Hybrid QRC)")
plt.title("Test Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```
